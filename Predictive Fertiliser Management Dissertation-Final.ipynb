```python
# Display the first few rows of each dataframe to understand their structure
(yield_data.head(), n2o_data.head(), ch4_data.head(), socsr_data.head())

# Function to handle missing values
def handle_missing_values(df, threshold=0.5):
    # Drop columns with more than 'threshold' proportion of missing values
    df = df.loc[:, df.isnull().mean() < threshold]
    
    # Impute missing numerical values with the median
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())
    
    # Impute missing categorical values with the mode
    cat_cols = df.select_dtypes(include=['object']).columns
    df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
    
    return df

# Apply the function to each dataset
yield_data_cleaned = handle_missing_values(yield_data)
n2o_data_cleaned = handle_missing_values(n2o_data)
ch4_data_cleaned = handle_missing_values(ch4_data)
socsr_data_cleaned = handle_missing_values(socsr_data)

# Check if there are any missing values left
(yield_data_cleaned.isnull().sum(), n2o_data_cleaned.isnull().sum(), ch4_data_cleaned.isnull().sum(), socsr_data_cleaned.isnull().sum())
```

```python
def split_data(data, target_column, test_size=0.3):
    X = data.drop(columns=[target_column])  # Features
    y = data[target_column]  # Target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    return X_train, X_test, y_train, y_test
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

def ella_update(task_data, num_latent_components, lambda_param, mu_param):
    max_features = max(X.shape[1] for X, _ in task_data)
    L = np.random.rand(num_latent_components, max_features)
    A = np.zeros((num_latent_components, num_latent_components))
    b = np.zeros((num_latent_components, max_features))

    for X, y in task_data:
        if X.shape[1] < max_features:
            X = np.pad(X, ((0, 0), (0, max_features - X.shape[1])), mode='constant')
        
        theta = np.linalg.inv(X.T @ X + lambda_param * np.eye(X.shape[1])) @ X.T @ y
        gbm_model = GradientBoostingRegressor(n_estimators=100, max_depth=3, learning_rate=0.1, random_state=42)
        LTX_T = L @ X.T
        gbm_model.fit(LTX_T.T, y)
        y_pred = gbm_model.predict(LTX_T.T)
        residuals = y - y_pred
        s = gbm_model.feature_importances_
        A += np.outer(s, s)
        theta_expanded = np.tile(theta, (num_latent_components, 1))
        b += s[:, np.newaxis] * theta_expanded
        L = np.linalg.inv(A + lambda_param * np.eye(A.shape[0])) @ b
    
    return L, A, b

def evaluate_ella(task_data, L, regularization_param=1e-5):
    performance = {}
    for X, y in task_data:
        L_sub = L[:, :X.shape[1]]
        transformed_X = X @ L_sub.T
        s = np.linalg.inv(transformed_X.T @ transformed_X + regularization_param * np.eye(transformed_X.shape[1])) @ transformed_X.T @ y
        y_pred = transformed_X @ s
        performance['mse'] = mean_squared_error(y, y_pred)
        performance['r2'] = r2_score(y, y_pred)
    return performance

def grid_search_ella(task_data, num_latent_components_list, lambda_param_list, mu_param_list):
    results = {}
    for i, (X, y) in enumerate(task_data):
        task_results = []
        for nc in num_latent_components_list:
            for lambda_param in lambda_param_list:
                for mu_param in mu_param_list:
                    L, A, b = ella_update([(X, y)], nc, lambda_param, mu_param)
                    performance = evaluate_ella([(X, y)], L)
                    mse = performance['mse']
                    r2 = performance['r2']
                    log_mse = np.log10(mse)
                    task_results.append((nc, lambda_param, mu_param, log_mse, r2))
                    print(f'Task {i+1}: nc={nc}, λ={lambda_param}, μ={mu_param} => log(MSE): {log_mse}, R²: {r2}')
        results[f'Task_{i+1}'] = task_results
    return results

def plot_r2_3d_per_task(results, lambda_param_list, mu_param_list):
    for task, task_results in results.items():
        unique_nc = set(r[0] for r in task_results)
        
        for nc in unique_nc:
            fig = plt.figure(figsize=(12, 5))
            
            # Extract data for plotting
            nc_results = [(lambda_param_list.index(lambda_param), mu_param_list.index(mu_param), r2) for nc_, lambda_param, mu_param, log_mse, r2 in task_results if nc_ == nc]
            lambda_indices, mu_indices, r2s = zip(*nc_results)
            
            # Plot R²
            ax2 = fig.add_subplot(111, projection='3d')
            ax2.scatter(lambda_indices, mu_indices, r2s, c=r2s, cmap='RdYlGn')
            ax2.plot_trisurf(lambda_indices, mu_indices, r2s, cmap='RdYlGn', alpha=0.8)
            ax2.set_xlabel('Lambda')
            ax2.set_ylabel('Mu')
            ax2.set_zlabel('R²')
            ax2.set_xlim(0, len(lambda_param_list) - 1)
            ax2.set_ylim(0, len(mu_param_list) - 1)
            ax2.set_zlim(min(r2s), max(r2s))
            ax2.set_xticks(range(len(lambda_param_list)))
            ax2.set_yticks(range(len(mu_param_list)))
            ax2.set_xticklabels(lambda_param_list)
            ax2.set_yticklabels(mu_param_list)
            ax2.set_title(f'{task} - R² for Latent Components = {nc}')
            
            # Highlight the highest R²
            max_r2 = max(r2s)
            max_index = r2s.index(max_r2)
            ax2.scatter(lambda_indices[max_index], mu_indices[max_index], max_r2, color='blue', s=100, label=f'Highest R²: {max_r2:.4f}')
            ax2.legend()

            # Add caption with the best hyperparameters
            caption = f"Best R² = {max_r2:.4f} for λ={lambda_param_list[lambda_indices[max_index]]}, μ={mu_param_list[mu_indices[max_index]]}"
            plt.figtext(0.5, -0.1, caption, ha="center", fontsize=12)
            
            plt.tight_layout()
            plt.show()

# Assuming task_data is prepared and task_data is defined as all_task_data
lambda_param_list = [1, 0.1, 0.01, 0.001, 0.0001]
mu_param_list = [1, 0.1, 0.01, 0.001, 0.0001]
num_latent_components_list = [10]

# Assuming you have all the task data in the following format
# all_task_data = [(yield_X_train, yield_y_train), (n2o_X_train, n2o_y_train), (ch4_X_train, ch4_y_train), (socsr_X_train, socsr_y_train)]

# Run grid search
results = grid_search_ella(all_task_data, num_latent_components_list, lambda_param_list, mu_param_list)

# Plot the R² in 3D for each task separately with captions
plot_r2_3d_per_task(results, lambda_param_list, mu_param_list)
```
